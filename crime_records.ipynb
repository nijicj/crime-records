```
from sklearn.preprocessing import StandardScaler
scaled_X = StandardScaler().fit_transform(X)             #normalize the values
scaled_X
```

```
type(scaled_X)
```

```
pd.DataFrame(scaled_X)    #convert numpy array to dataframe
```

```
pd.DataFrame(scaled_X).describe()
```

```
features = scaled_X.T                       #transpose of matrix
covariance_matrix = np.cov(features)      #find covarience matrix
print(covariance_matrix)
```

```
covariance_matrix.shape
```

```
eig_vals, eig_vecs = np.linalg.eig(covariance_matrix)  #Eigen values and eigen vectors calculated
print(eig_vals)
```

```
print(eig_vecs)
```

```
eig_vals[0] / sum(eig_vals)
```

```
sum(eig_vals)
```

```
eig_vals[1] / sum(eig_vals)
```

```
eig_vals[2] / sum(eig_vals)
```

```
eig_vals[3] / sum(eig_vals)
```

```
eig_vecs.shape
```

```
V = eig_vecs[:, :2]      #selecting two PCA
```

```
V.shape
```

```
projected_X = scaled_X.dot(V)        #new matrix of PC1 and PC2
projected_X
```

```
type(projected_X)
```

```
X_pca = pd.DataFrame(data = projected_X, columns = ['PC1', 'PC2'])
X_pca.head()
```

```
X_pca.shape
```

```
plt.figure(figsize=(5,4))
plt.scatter(X_pca['PC1'], X_pca['PC2'])
plt.xlabel('First principle component')
plt.ylabel('Second principle component')
```

```
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=5)
kmeans.fit(X_pca)
```

```
labels = kmeans.predict(X_pca)
centroids = kmeans.cluster_centers_
```
